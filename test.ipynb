```python
# coverage_risk
def coverage_risk(confidence, sort_correctness):
    confidence = sort_softmax_max
    sort_correctness

    risk_list = []
    coverage_list = []
    risk = 0

    for i in range(len(confidence)):
        # from 0 to 1
        coverage = (i + 1) / len(confidence)
        coverage_list.append(coverage)
        
        # wrong prediction
        # 错误预测占总预测的比例
        if sort_correctness[i] == 0:
            risk += 1
        
        risk_list.append(risk / (i + 1))
        
    return risk_list, coverage_list
risk_li, coverage_li = coverage_risk(sort_softmax_max, sort_correctness)
```

```python

# 利用错误率累计曲线计算
def aurc_eaurc(risk_list):
    risk_list = risk_li
    # r is the total mistake rate
    r = risk_list[-1]
    print(f'risk_list: {risk_list}')
    risk_coverage_curve_area = 0

    # is a threshold
    optimal_risk_area = r + (1 - r) * np.log(1 - r)

    for risk in risk_list:
        risk_coverage_curve_area += risk * (1 / len(risk_list))
        

    aurc = risk_coverage_curve_area
    eaurc = risk_coverage_curve_area - optimal_risk_area

    print(f'aurc: {aurc}')
    print(f'eaurc: {eaurc}')

    print("AURC {0:.2f}".format(aurc*1000))
    print("EAURC {0:.2f}".format(eaurc*1000))
    return aurc, eaurc

```

```python

# calc fpr auroc aupr 

def calc_fpr_aupr(list_softmax, list_correct):
    softmax_np = np.array(list_softmax)
    # print(f'softmax_np: {softmax_np.shape}')
    correctness_np = np.array(list_correct)
    # 得到每个softmax的最大值
    softmax_max_np = np.max(softmax_np, 1)
    # print(f'softmax_max_np: {softmax_max_np.shape}')
    from sklearn import metrics

    # 顺序排列后，逐个预测，计算fpr和tpr， 从socre中得到的是预测变化
    fpr, tpr, thresholds = metrics.roc_curve(correctness_np, softmax_max_np)
    print(f'fpr: {fpr.shape}')
    print(f'tpr: {tpr.shape}')
    print(f'thresholds: {thresholds.shape}')
    auroc = metrics.auc(fpr, tpr)

    idx_tpr_95 = np.argmin(np.abs(tpr - 0.95))
    print(f'idx_tpr_95: {idx_tpr_95}')
    fpr_in_tpr_95 = fpr[idx_tpr_95]
    print(f'fpr_in_tpr_95: {fpr_in_tpr_95}')
    tpr_in_tpr_95 = tpr[idx_tpr_95]
    print(f'tpr_in_tpr_95: {tpr_in_tpr_95}')

    precision, recall, thresholds = metrics.precision_recall_curve(correctness, softmax_max)
    aupr_success = metrics.auc(recall, precision)

    aupr_err = metrics.average_precision_score(-1 * correctness + 1, -1 * softmax_max)

    print("AUROC {0:.2f}".format(auroc * 100))
    print('AUPR_Success {0:.2f}'.format(aupr_success * 100))
    print("AUPR_Error {0:.2f}".format(aupr_err*100))
    print('FPR@TPR95 {0:.2f}'.format(fpr_in_tpr_95*100))

    return auroc, aupr_success, aupr_err, fpr_in_tpr_95

auroc, aupr_success, aupr_err, fpr_in_tpr_95 = calc_fpr_aupr(list_softmax, list_correct)
```

```python
import numpy as np
import pandas as pd
import os
from sklearn import metrics


import torch
import pickle
import torch.nn.functional as F
#  Reliability Diagrams

f = open('demo.pckl', 'rb')
output, _, label = pickle.load(f)
softmax = F.softmax(output)

softmax = torch.tensor(softmax)
label = torch.tensor(label)

# print(softmax)
bins = 15

bin_boundaries = torch.linspace(0, 1, bins + 1)
bin_lowers = bin_boundaries[:-1]
bin_uppers = bin_boundaries[1:]


msp, predictions = torch.max(softmax, 1)
correctness = predictions.eq(label)
```

```python
ece = torch.zeros(1)
for bin_lower, bin_upper in zip(bin_lowers, bin_uppers):
    in_bin = msp.gt(bin_lower.item()) * msp.le(bin_upper.item())
    prop_in_bin = in_bin.float().mean()
    if prop_in_bin.item() > 0:
        accuracy_in_bin = correctness[in_bin].float().mean()
        avg_confidence_in_bin = msp[in_bin].mean()
        ece += torch.abs(avg_confidence_in_bin - accuracy_in_bin) * prop_in_bin
```

```python
from data.dataset import *
import torch
from torchvision.datasets import CIFAR100
class CIFAR100_loss(CIFAR100):
    def __init__(
        self,
        root: str,
        train: bool = True,
        transform: Optional[Callable] = None,
        target_transform: Optional[Callable] = None,
        download: bool = False,
        loss_dir : str = None,
        cut_rate: float = 0.0,
    ) -> None:

        super().__init__(root, transform=transform, target_transform=target_transform)
        
        self.losses = None
        if loss_dir is not None:
            self.losses = []
            with open(loss_dir, 'rb') as f:
                self.losses = pickle.load(f)
        
            if cut_rate > 0:
                # cut the loss by the value of loss, drop the least loss
                print('use cut_rate: ', cut_rate)
                loss_sort = np.sort(self.losses)
                cut_index = int(len(loss_sort) * cut_rate)
                self.save_index = [self.losses.index(value) for value in loss_sort[cut_index:]]
                self._refresh_data()
                
    def _refresh_data(self):
        self.data = [self.data[i] for i in self.save_index]
        self.targets = [self.targets[i] for i in self.save_index]
        # if self.losses is not None:
        self.losses = [self.losses[i] for i in self.save_index]
        
        print('refresh data', len(self.data), len(self.targets), len(self.losses))
                
            
    
    def __getitem__(self, index: int) -> Tuple[Any, Any]:
        """
        Args:
            index (int): Index

        Returns:
            tuple: (image, target, loss) where target is index of the target class.
        """
        
        img, target = self.data[index], self.targets[index]

        img = Image.fromarray(img)
        # img.save('./temp.jpg')
        # print('save img', img.size, img.mode, target)
        # doing this so that it is consistent with all other datasets
        # to return a PIL Image
        # img = Image.fromarray(img)

        if self.transform is not None:
            img = self.transform(img)
        # print('target', target)
        if self.target_transform is not None:
            target = self.target_transform(target)
            # print('target', target)
        if self.losses is not None:
            loss = self.losses[index]

            return img, target, loss
        else:
            return img, target
```
